Import the libraries that are required for data analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SampleSuperstore.csv')
print("Data imported!!")
```

### **Obesrvation 1**

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue', 'orange'], figsize=(7,5))
plt.ylabel('Profit Loss')
plt.show()
```

Profit is high in West region and low in Central region. While sales is high in West region and low in South region

### **Obesrvation 2**

```python
df.groupby('Segment')[['Profit', 'Sales']].sum().plot.bar(color=['brown', 'orange'], figsize=(7, 5))
plt.ylabel('Profit Loss')
plt.show()
```

Consumer segment has most sales and profit whereas Home Office segment has least profit and sales

### **Obesrvation 3**

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=df,palette='coolwarm_r',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

California has the highest buyers across all states

### **Obesrvation 4**

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['green','orange'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit&Loss and Sales')
plt.show()
```

Technology category has the highest profit and sales altogether and furniture has the lowest profit

### **Obesrvation 5**

```python
sns.lineplot(x='Discount',y='Profit', label='Profit', data=df, color='orange')
```

No correlation between profit and discount

### **Obesrvation 6**

```python
df.hist(figsize=(10,10), bins=50, color='orange')
plt.show()
```

2 and 3 are most preferred quantity customers buy and 0-20% is the discount given upon purchase. While profit remains near to zero.
